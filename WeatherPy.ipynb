```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
cities
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# sets base url 
url = "http://api.openweathermap.org/data/2.5/weather?"
# creates a dictionary for storing information
city_dict = {
"City": [],
"Latitude": [],
"Longitude":[],
"Temperature": [],
"Humidity": [],   
"Cloudiness": [],
"Wind Speed": [],
"Country": [],
"Date": []
} 
# vatiable for tracking city #
x = 0
# grabs infromation from weather database and stores it in teh dictionary 
for city in cities:
    city_url = url + "appid=" + weather_api_key + "&q=" + city
    city_json = requests.get(city_url).json()
    try:
         # shows that the a city has been processed
        print(f"processing{city}, Number {x} of {len(cities)}" )
        city_dict["City"].append(city_json["name"])
        city_dict["Latitude"].append(city_json["coord"]["lat"])
        city_dict["Longitude"].append(city_json["coord"]["lon"])
        city_dict["Temperature"].append(city_json["main"]["temp_max"])
        city_dict["Humidity"].append(city_json["main"]["humidity"])
        city_dict["Cloudiness"].append(city_json["clouds"]["all"])
        city_dict["Wind Speed"].append(city_json["wind"]["speed"])
        city_dict["Country"].append(city_json["sys"]["country"])
        city_dict["Date"].append(city_json["sys"]["sunrise"])
        x += 1
    # shows if a city fails to be processed
    except (KeyError,IndexError):
        print("not found")
        x += 1
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# converts dictionary to a data frame and checks how many cities are in it
city_df = pd.DataFrame(city_dict)
city_df.count()
```

## Linear Regression

```python
# Creates Northern and Southern Hemisphere DataFrames
NorthH = city_df[city_df["Latitude"] >= 0]
SouthH = city_df[city_df["Latitude"] < 0]
print(NorthH.head())
print(SouthH.head())
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# plots and saves a graph of the Northern Hemisphere Temp vs. Latitude Linear Regression
x_values = NorthH["Latitude"]
y_values = NorthH["Temperature"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq, xy=(50,310),fontsize=15,color="red")
plt.xlabel('Latitude')
plt.ylabel('Temperature')
plt.title("Northern Hemisphere - Temperature vs. Latitude")
print(f"The r-squared is: {rvalue**2}")
plt.savefig('output_data/NH_Lat_Temp_plt.png')
print("This plot shows that there is strong relationship between the latitude and temperature of city in the Northern Hemisphere.")
```
